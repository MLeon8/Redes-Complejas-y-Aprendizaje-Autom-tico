```python
import networkx as nx

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from scipy import *

import math as math
import seaborn as sns

# %matplotlib inline

from ipywidgets import *

anch = 20
alt = anch/float(1.6180334)

 

```

```python


BA = nx.barabasi_albert_graph(22000,10)
fig, ax = plt.subplots(figsize=(anch, alt))

nx.draw(BA)


dBA = dict(nx.degree(BA))
seqBA = sorted(dBA.values())
#nx.draw_spring(RCB) #Solo nodos
#nx.draw_networkx(RCB) #Si se quiere saber el n√∫mero del nodo
#nx.draw_kamada_kawai(RCB)  #Kamada Kawai (es mas lento)

#Kavg = sum(dBA.values()) / len(dBA)  
cc = nx.average_clustering(BA)
spl = nx.average_shortest_path_length(BA)
k = 20

```

```python


RCB = nx.barabasi_albert_graph(22000,10)
    
dRCB = dict(nx.degree(RCB))
seqRCB = sorted(dBA.values())    
    
Kavg = sum(seqRCB) / len(dRCB)  
cc = nx.average_clustering(RCB)
spl = nx.average_shortest_path_length(RCB)

cajas = 100
histRCB = plt.hist(seqBA, bins=cajas)

K = []
for i in range(len(histRCB[1])-1):
    K.append(histRCB[1][i])

```

```python


fig, ax = plt.subplots(figsize=(anch, alt))

sns.histplot(seqRCB,stat='count',kde=True, fill= True, element='step', label='$< k > =$ %f \n $< C > =$ %f \n $ < l > =$ %f' % (Kavg,cc,spl))


plt.title("Distribuci√≥n de grado. \n Modelo de Barabasi-Alberts")

plt.xlabel('$k$')
plt.ylabel('$P(k)$')
plt.legend()

```

```python
fig, ax = plt.subplots(figsize=(anch, alt))

plt.loglog(K,histRCB[0], 'r.', label='$< k > =$ %f \n $< C > =$ %f \n $ < l > =$ %f' % (Kavg,cc,spl))

plt.title("Distribuci√≥n de grado en escala logaritmica. \n Modelo de Barabasi-Alberts")
plt.xlabel('$log(k)$')
plt.ylabel('$log(P(k))$')

plt.legend()
```
